```python
import pandas as pd
import numpy as np
import random

# -----------------------------
# CONFIGURATION
# -----------------------------
NUM_RECORDS = 1000
np.random.seed(42)

teachers = ["Mr. Sharma", "Ms. Patel", "Mr. Khan", "Ms. Iyer"]
sections = ["A", "B", "C"]
subjects = ["Math", "Science", "English", "Computer"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

# -----------------------------
# DATA GENERATION
# -----------------------------
data = []

for _ in range(NUM_RECORDS):
    score = np.random.randint(35, 100)
    attendance = np.random.randint(60, 100)
    late_count = np.random.randint(0, 6)

    performance_percent = score

    risk_level = "High Risk" if (
        performance_percent < 50 or
        attendance < 70 or
        late_count > 3
    ) else "Normal"

    data.append([
        np.random.randint(1000, 5000),       # student_id
        random.choice(teachers),
        random.choice(sections),
        random.choice(subjects),
        score,
        attendance,
        late_count,
        random.choice(["Active", "Dropped"],),
        random.choice(months),
        performance_percent,
        risk_level
    ])

columns = [
    "student_id",
    "teacher",
    "section",
    "subject",
    "score",
    "attendance_percent",
    "late_count",
    "status",
    "month",
    "performance_percent",
    "risk_level"
]

df = pd.DataFrame(data, columns=columns)

df.to_csv("advanced_academic_performance_dataset.csv", index=False)

print("Dataset generated successfully!")
```
